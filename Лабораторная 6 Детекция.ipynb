```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("yolo11x.pt")
```

```python
input_video_path = '/kaggle/input/coca-cola-add/Coca Cola Secret Santa.mp4'
output_video_path = 'cola_detection_video.mp4'
```

```python
cap = cv2.VideoCapture(input_video_path)
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Детектирование объектов на текущем кадре
    results = model(frame)

    # Наносим аннотации на кадр
    annotated_frame = results[0].plot()  # Генерация аннотированного кадра

    # Запись кадра в выходное видео
    out.write(annotated_frame)

# Закрытие ресурсов
cap.release()
out.release()
```
